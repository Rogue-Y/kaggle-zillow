```python
for col in prop2017.columns:
    print('-----------------------------------------------')
    print('2017 %s:' %col)
    print(prop2017[col].describe())
    print('\n2016 %s:' %col)
    print(prop2016[col].describe())
    print('-----------------------------------------------\n')
```

```python
for col in prop2017.columns:
    print('-----------------------------------------------')
    print('2017 %s count:' %col)
    print(prop2017[col].count())
    print('\n2016 %s count:' %col)
    print(prop2016[col].count())
    print('-----------------------------------------------\n')
```

```python
# All columns where 2017 has more data
i = 0
for col in prop2017.columns:
    if prop2017[col].count() > prop2016[col].count():
        i+=1
        print('-----------------------------------------------')
        print('2017 %s count: %d' %(col, prop2017[col].count()))
        print('\n2016 %s count: %d' %(col, prop2016[col].count()))
        print('-----------------------------------------------\n')

print('Total columns with more data: %d' %i)
```

```python
# All columns where 2017 has less data
i = 0
for col in prop2017.columns:
    if prop2017[col].count() < prop2016[col].count():
        i+=1
        print('-----------------------------------------------')
        print('2017 %s count: %d' %(col, prop2017[col].count()))
        print('\n2016 %s count: %d' %(col, prop2016[col].count()))
        print('-----------------------------------------------\n')

print('Total columns with less data: %d' %i)
```
