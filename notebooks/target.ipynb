```python
import pandas as pd
```

```python
prop = TR.prepare_features(feature_list)
train_df = utils.load_transaction_data()
df = train_df.merge(prop, how='left', on='parcelid')
```

```python
def aggregate_by_region(df, id_name, column='logerror'):
    region_dict = (df[[id_name, column]].groupby(id_name)
        .agg(['max', 'min', 'std', 'mean']).to_dict())
    default_value_dict=df[[column]].agg(['max', 'min', 'std', 'mean']).to_dict()[column]
    return region_dict
    
```

```python
rd = aggregate_by_region(df, 'regionidcity')
```

```python
def generate_target_feature(df, id_name, region_dict):
    newdf = pd.DataFrame()
    target = list(rd.keys())[0][0]
    stats = [item[1] for item in rd.keys()]
    for stat in stats:
        newdf[target + '_' + id_name + '_' + stat] = df[id_name].map(region_dict[target, stat])
    return newdf
```
